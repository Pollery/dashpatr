# APARENTEMENTE INUTIL


```python
import pandas as pd
```

```python
df_faturas = pd.read_csv("../data/consolidado/df_faturas.csv")
```

```python
df_faturas["date"] = pd.to_datetime(df_faturas["date"], format="%Y-%m-%d")
df_faturas.head()
```

```python
df_faturas["Month"] = df_faturas["date"].dt.month
df_faturas["Year"] = df_faturas["date"].dt.year
```

```python
# Valores
pd.DataFrame(df_faturas.groupby(["Year","Month"])["amount"].sum())
```

```python
# Estabelecimento
pd.DataFrame(df_faturas["title"].value_counts()).head()
```

```python

pd.DataFrame(df_faturas.groupby(["title"])["amount"].sum().sort_values(ascending=False)).head(5)    
```
